````python
import os
from dotenv import load_dotenv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
from openai import OpenAI
from prompts import question_prompt, assessment_prompt
# Load environment variables
load_dotenv()


# Access email and password from .env file
email_id = os.getenv("EMAIL")
password = os.getenv("PASSWORD")

def login():
    # Wait until the email input field is present
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '/html/body/div/section/section/main/div/div[1]/main/form[1]/div[1]/div[1]/input'))
    )

    # Find the email input field and enter the email
    email_input = driver.find_element(By.XPATH, '/html/body/div/section/section/main/div/div[1]/main/form[1]/div[1]/div[1]/input')
    email_input.send_keys(email_id)

    # Find the password input field and enter the password
    password_input = driver.find_element(By.XPATH, '/html/body/div/section/section/main/div/div[1]/main/form[1]/div[2]/div[1]/input')
    password_input.send_keys(password)

    # Submit the form
    password_input.send_keys(Keys.RETURN)
    

def click_optional_buttons():
    # optional button 1
    # Click the specified button
    try:
        button = driver.find_element(By.XPATH, '/html/body/div/section/section/main/div/main/div[1]/div/div[1]/div[1]/div/div[3]/button')
        button.click()
    except:
        pass
    try:
        # optional button 2
        # Click the got it button
        button = driver.find_element(By.XPATH, '/html/body/div/div[1]/section[1]/div[2]/div/button')
        button.click()
    except:
        pass
    try:
        # optional button 3
        # Click the got it button
        button = driver.find_element(By.XPATH, '/html/body/div/div[1]/section[2]/div[1]/div/button')
        button.click()
    except:
        pass
    try:
        # optional button 4
        # Click the "Take the Level 2 Assessment" Button
        button = driver.find_element(By.XPATH, '/html/body/div/section/section/main/div/main/nav/div/button')
        button.click()
    except:
        pass
    try:
        # optional button 5
        # Click the "Submit Assessment" Button
        button = driver.find_element(By.XPATH, '/html/body/div/div[1]/div/section/div[2]/footer/button[1]')
        button.click()
    except:
        pass
    try:
        # optional button 6
        # Click the "Continue" Button after completing the assessment
        button = driver.find_element(By.XPATH, '/html/body/div/section/section/main/div/div/div[1]/button')
        button.click()
    except:
        pass
    try:
        # optional button 7
        # Click the "No thanks" Button for ads
        button = driver.find_element(By.XPATH, '/html/body/div/div[1]/div/section/div[2]/main/div/div[2]/button[2]')
        button.click()
    except:
        pass
    try:
        # optional button 8
        # Click the Video Play button
        button = driver.find_element(By.XPATH, '/html/body/div/section/section/main/main/nav/div/div[3]/button')
        button.click()
    except:
        pass
    try:
        # optional button 9
        # Click the Video play button
        button = driver.find_element(By.XPATH, '/html/body/div/section/section/main/main/article/div/div[1]/div/button')
        button.click()
    except:
        pass
    
    

def check_question_section():
    # Check if the specific section exists
    try:
        section = driver.find_element(By.XPATH, '/html/body/div/section/section/main/main/article/div')
        if section.get_attribute("data-test") == "popupQuestionContainer":
            print(True)
            print("The Question section exists on the page.")
            question_section_html = section.get_attribute('outerHTML')
            return question_section_html
        else:
            pass
    except:
        pass
    return None

def check_assessment_section():
    try:
        section = driver.find_element(By.XPATH, '/html/body/div/section/section/main/div/article/div')
        if section.get_attribute("class") == "gritCourseflowNode__question-wrapper":
            print(True)
            print("The Question section exists on the page.")
            assessment_section_html = section.get_attribute('outerHTML')
            return assessment_section_html
        else:
            pass
    except:
        pass
    return None

def answer_assessment(answer_text):
    try:
        escaped_text = answer_text.replace("'", "\'")
    
        # Use an XPath with normalize-space() and contains() to handle multiline and special characters
        button = driver.find_element(By.XPATH, f"//button[contains(normalize-space(), \"{escaped_text}\")]")
        button.click()
        print(f"Clicked the button with text {answer_text}.")
        try:
            submit_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '/html/body/div/section/section/main/div/nav/div/div[3]/button'))
            )
            submit_button.click()
            print("Clicked on the submit button.")
        except Exception as e:
            print(f"Error occured while clicking Submit. {e}")
    except Exception as e:
        print(f"An error occurred: {e}")

def answer_question(answer_button_Xpath):
    try:
        answer_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, answer_button_Xpath))
        )
        answer_button.click()
        print("Clicked on the answer button.")
        try:
            submit_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '/html/body/div/section/section/main/main/nav/div/div[3]/button'))
            )
            submit_button.click()
            print("Clicked on the submit button.")
            answer_status = answer_status = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH,'/html/body/div/section/section/main/main/article/div/div[2]/div/div[1]/div[2]/div/div/h3'))
                ).text
            return answer_status
        except Exception as e:
            print(f"An error occurred while clicking the submit button: {e}")
    except Exception as e:
        print(f"An error occurred while clicking the answer button: {e}")
    return None

def click_continue_learning_button(ul_xpath):
    """
    Finds and clicks the continue learning button by scanning through the ul element structure.
    
    Parameters:
        ul_xpath (str): The XPath to the ul element containing the course list items.
    
    Returns:
        bool: True if the button was found and clicked, False otherwise.
    """
    try:
        # Find the ul element
        ul_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, ul_xpath))
        )
        
        # Get all li elements within the ul
        li_elements = ul_element.find_elements(By.TAG_NAME, "li")
        
        # Iterate through each li element
        for li in li_elements:
            try:
                # Find the button within the li element
                button = li.find_element(By.TAG_NAME, "button")
                
                # Find all div elements within the button
                divs = button.find_elements(By.TAG_NAME, "div")
                
                # Scan through the divisions for class "continue-learning-button"
                for div in divs:
                    class_attr = div.get_attribute("class")
                    if class_attr and "continue-learning-button" in class_attr:
                        # Found the div with continue-learning-button class
                        # Now find the button with data-test="continueLearningButton" inside this div
                        try:
                            continue_button = div.find_element(By.CSS_SELECTOR, 'button[data-test="continueLearningButton"]')
                            continue_button.click()
                            print("Successfully clicked the continue learning button.")
                            return True
                        except Exception as e:
                            print(f"Could not find or click the continueLearningButton: {e}")
                            continue
            except Exception as e:
                # If no button found in this li, continue to next li
                continue
        
        print("Continue learning button not found in any course item.")
        return False
        
    except Exception as e:
        print(f"An error occurred while searching for the continue learning button: {e}")
        return False

def click_continue_learning_button_direct():
    """
    Finds and clicks the button with data-test="continueLearningButton" directly.
    
    Returns:
        bool: True if the button was found and clicked, False otherwise.
    """
    try:
        # Find the button with data-test="continueLearningButton"
        continue_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'button[data-test="continueLearningButton"]'))
        )
        continue_button.click()
        print("Successfully clicked the continue learning button.")
        return True
    except Exception as e:
        print(f"An error occurred while finding or clicking the continue learning button: {e}")
        return False

def click_landing_page_continue_button():
    """
    Finds and clicks the button with data-test="landingPageContinueButton".
    
    Returns:
        bool: True if the button was found and clicked, False otherwise.
    """
    try:
        # Find the button with data-test="landingPageContinueButton"
        landing_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'button[data-test="landingPageContinueButton"]'))
        )
        landing_button.click()
        print("Successfully clicked the landing page continue button.")
        return True
    except Exception as e:
        print(f"An error occurred while finding or clicking the landing page continue button: {e}")
        return False

def click_courseflow_forward():
    """
    Finds and clicks the button with data-test="courseflow-forward".
    
    Returns:
        bool: True if the button was found and clicked, False otherwise.
    """
    try:
        # Find the button with data-test="courseflow-forward"
        forward_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'button[data-test="courseflow-forward"]'))
        )
        forward_button.click()
        print("Successfully clicked the courseflow forward button.")
        return True
    except Exception as e:
        print(f"An error occurred while finding or clicking the courseflow forward button: {e}")
        return False

def click_courseflow_back():
    """
    Finds and clicks the button with data-test="courseflow-back".
    
    Returns:
        bool: True if the button was found and clicked, False otherwise.
    """
    try:
        # Find the button with data-test="courseflow-back"
        back_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'button[data-test="courseflow-back"]'))
        )
        back_button.click()
        print("Successfully clicked the courseflow back button.")
        return True
    except Exception as e:
        print(f"An error occurred while finding or clicking the courseflow back button: {e}")
        return False

def click_play_button():
    """
    Finds and clicks the button with data-test="playButton".
    
    Returns:
        bool: True if the button was found and clicked, False otherwise.
    """
    try:
        # Find the button with data-test="playButton"
        play_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'button[data-test="playButton"]'))
        )
        play_button.click()
        print("Successfully clicked the play button.")
        return True
    except Exception as e:
        print(f"An error occurred while finding or clicking the play button: {e}")
        return False

def extract_course_container_content():
    """
    Extracts content from the element with data-test="courseContainer" and returns it
    as a compact markdown-formatted string.
    
    Returns:
        str: The content in markdown format, or None if element not found.
    """
    def html_to_markdown(soup_element):
        """Recursively convert HTML elements to markdown format."""
        if soup_element is None:
            return ""
        
        result = []
        
        for element in soup_element.children:
            if hasattr(element, 'name') and element.name is not None:  # It's a tag
                tag_name = element.name.lower()
                try:
                    text = element.get_text(strip=True) if not element.find_all() else ""
                    text = text if text is not None else ""
                except:
                    text = ""
                
                if tag_name in ['script', 'style', 'noscript']:
                    continue
                elif tag_name == 'h1':
                    result.append(f"# {text}\n")
                elif tag_name == 'h2':
                    result.append(f"## {text}\n")
                elif tag_name == 'h3':
                    result.append(f"### {text}\n")
                elif tag_name == 'h4':
                    result.append(f"#### {text}\n")
                elif tag_name == 'h5':
                    result.append(f"##### {text}\n")
                elif tag_name == 'h6':
                    result.append(f"###### {text}\n")
                elif tag_name == 'p':
                    para_text = html_to_markdown(element).strip()
                    if para_text:
                        result.append(f"{para_text}\n\n")
                elif tag_name == 'strong' or tag_name == 'b':
                    result.append(f"**{text}**")
                elif tag_name == 'em' or tag_name == 'i':
                    result.append(f"*{text}*")
                elif tag_name == 'code':
                    result.append(f"`{text}`")
                elif tag_name == 'pre':
                    result.append(f"```\n{text}\n```\n")
                elif tag_name == 'a':
                    href = element.get('href', '') or ''
                    link_text = html_to_markdown(element).strip()
                    if href:
                        result.append(f"[{link_text}]({href})")
                    else:
                        result.append(link_text)
                elif tag_name == 'img':
                    src = element.get('src', '') or ''
                    alt = element.get('alt', '') or ''
                    result.append(f"![{alt}]({src})")
                elif tag_name == 'ul':
                    result.append("\n")
                    for li in element.find_all('li', recursive=False):
                        li_text = html_to_markdown(li).strip()
                        if li_text:
                            result.append(f"- {li_text}\n")
                    result.append("\n")
                elif tag_name == 'ol':
                    result.append("\n")
                    for idx, li in enumerate(element.find_all('li', recursive=False), 1):
                        li_text = html_to_markdown(li).strip()
                        if li_text:
                            result.append(f"{idx}. {li_text}\n")
                    result.append("\n")
                elif tag_name == 'li':
                    # Handle nested content
                    nested = html_to_markdown(element)
                    result.append(nested)
                elif tag_name == 'br':
                    result.append("\n")
                elif tag_name == 'hr':
                    result.append("\n---\n")
                elif tag_name == 'blockquote':
                    quote_text = html_to_markdown(element).strip()
                    if quote_text:
                        result.append(f"> {quote_text.replace(chr(10), chr(10) + '> ')}\n\n")
                elif tag_name == 'table':
                    result.append("\n")
                    rows = element.find_all('tr', recursive=False)
                    if rows:
                        # Header row
                        header = rows[0]
                        headers = [th.get_text(strip=True) for th in header.find_all(['th', 'td'], recursive=False)]
                        if headers:
                            result.append("| " + " | ".join(headers) + " |\n")
                            result.append("| " + " | ".join(["---"] * len(headers)) + " |\n")
                        # Data rows
                        for row in rows[1:]:
                            cells = [td.get_text(strip=True) for td in row.find_all(['td', 'th'], recursive=False)]
                            if cells:
                                result.append("| " + " | ".join(cells) + " |\n")
                    result.append("\n")
                else:
                    # For other tags, recursively process children
                    nested = html_to_markdown(element)
                    result.append(nested)
            else:
                # It's a text node
                try:
                    text = str(element).strip() if element is not None else ""
                    if text:
                        result.append(text)
                except:
                    pass
        
        return "".join(result)
    
    try:
        # Find the element with data-test="courseContainer"
        container = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, '[data-test="courseContainer"]'))
        )
        
        # Get the HTML content
        html_content = container.get_attribute('outerHTML')
        
        # Parse with BeautifulSoup
        soup = BeautifulSoup(html_content, 'html.parser')
        
        # Remove script and style elements
        for script in soup(["script", "style"]):
            script.decompose()
        
        # Convert to markdown
        markdown_content = html_to_markdown(soup)
        
        # Clean up extra whitespace
        lines = markdown_content.split('\n')
        cleaned_lines = []
        prev_empty = False
        for line in lines:
            stripped = line.strip()
            if not stripped:
                if not prev_empty:
                    cleaned_lines.append('')
                    prev_empty = True
            else:
                cleaned_lines.append(line)
                prev_empty = False
        
        result = '\n'.join(cleaned_lines).strip()
        print(f"Successfully extracted course container content ({len(result)} characters).")
        return result
        
    except Exception as e:
        print(f"An error occurred while extracting course container content: {e}")
        return None
````

````python
from openai import OpenAI

def get_gpt4o_response(prompt, model="gpt-4o", temperature=0, max_tokens=300):
    """
    Function to send a prompt to the OpenAI GPT-4 model and retrieve a response.

    Parameters:
        prompt (str): The input prompt to send to the GPT model.
        model (str): The model to use (default is "gpt-4").
        temperature (float): Sampling temperature. Higher values make the output more random.
        max_tokens (int): The maximum number of tokens to include in the response.

    Returns:
        str: The model's response to the input prompt.
    """
    try:
        
        client = OpenAI(
            api_key=os.getenv("OPENAI_API_KEY")
        )
        # Send the prompt to the OpenAI API
        response = client.chat.completions.create(
            messages=[{"role": "user", "content": prompt,}],
            temperature=temperature,
            model=model,
        )

        # Extract and return the response content
        return response.choices[0].message.content
    except Exception as e:
        return f"An error occurred: {str(e)}"
````
